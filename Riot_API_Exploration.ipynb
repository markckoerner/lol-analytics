```python
import requests
```

API Docs: [Link](https://developer.riotgames.com/apis)

```python
# get match list from summoner_name
api_key = 'RGAPI-62677d21-aca0-4fac-b84c-64126f0b6676' #only valid 24h
summoner_name = 'welikefn'

#get summoner accountID
resp_sn = requests.get('https://euw1.api.riotgames.com/lol/summoner/v4/summoners/by-name/'+summoner_name+'?api_key='+api_key)
if resp_sn.status_code != 200:
    # This means something went wrong.
    print('Summoner API Error. Error Code: ' + str(resp_sn.status_code))
accID = resp_sn.json()['accountId']

#get match list
resp_ml = requests.get('https://euw1.api.riotgames.com/lol/match/v4/matchlists/by-account/'+accID+'?api_key='+api_key)
if resp_ml.status_code != 200:
    # This means something went wrong.
    print('Matches API Error. Error Code: ' + str(resp_sn.status_code))
matches = resp_ml.json()['matches']
totalGames = resp_ml.json()['totalGames']
endIndex = resp_ml.json()['endIndex']

while endIndex < totalGames:
    begIndex = str(endIndex)
    resp_ml_cont = requests.get('https://euw1.api.riotgames.com/lol/match/v4/matchlists/by-account/'+accID+'?beginIndex='+begIndex+'&api_key='+api_key)
    if resp_ml_cont.status_code != 200:
        # This means something went wrong.
        print('Matches API Error. Error Code: ' + str(resp_ml_cont.status_code))
    matches.extend(resp_ml_cont.json()['matches'])
    totalGames = resp_ml_cont.json()['totalGames']
    endIndex = resp_ml_cont.json()['endIndex']
    print(endIndex, totalGames, len(matches))
```

```python
gameIdList = [dic['gameId'] for dic in matches]
gameIdList[:5]
```
